```python
5.Construir una tabla que muestre cuántas medallas obtuvieron los hombres en total en cada año que se realizó el evento.
```

```python
df_Summer_Olympic_medals_men = df_Summer_Olympic_medals[df_Summer_Olympic_medals['Gender'].str.contains("Men")]
df_Summer_Olympic_medals_men
```

```python
df_Summer_Olympic_medals_men.describe()
```

```python
#https://interactivechaos.com/es/manual/tutorial-de-pandas/agrupaciones-en-dataframes
df_Summer_Olympic_medals_men_Year= df_Summer_Olympic_medals_men.groupby(by = ['Year']).count()
df_Summer_Olympic_medals_men_Year['Medal'] # nro de medallas x Year para los Hombres
```

```python
df_Summer_Olympic_medals_Gender= df_Summer_Olympic_medals.groupby(by = ['Year','Gender']).count()
df_Summer_Olympic_medals_Gender['Medal'] # nro de medallas x Year para los Hombres
```
