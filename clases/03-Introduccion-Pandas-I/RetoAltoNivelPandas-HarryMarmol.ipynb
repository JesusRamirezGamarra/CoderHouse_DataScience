***8. Muestra el porcentaje de personas que sobrevivieron y murieron.***


```python
#Creamos un dataframe con la informacion requerida
porcentajes = df_titanic.groupby("Survived").agg({"Survived":'count'})
#Aplicamos una formula para calcular el porcentaje al nuevo DF
porcentajes['%'] = (porcentajes['Survived'] / 
                  porcentajes['Survived'].sum()) * 100
porcentajes

```

***9. Muestra el porcentaje de personas que sobrevivieron en cada clase.***

```python
#Creamos un dataset con la informacion agrupada
ppc = df_titanic.groupby(["Survived","Pclass"]).agg({"PassengerId":'count'})
#Aplico una formula de porcentaje para este nuevo dataframe con las columnas agrupadas
ppc["%"] = (ppc['PassengerId'] / 
                  ppc['PassengerId'].sum()) * 100
ppc
```

***10. Elimina del DataFrame los pasajeros con edad desconocida.***

```python
#Eliminamos los NaN
df_titanic = df_titanic.dropna(subset=['Age'])
#Reseteamos el indice
df_titanic = df_titanic.reset_index(drop=True)
#Mostramos el dataset
df_titanic
```

```python
#Creamos un dataframe con la edad media aplicando agrupacion y agregacion
edad_media = df_titanic.groupby("Pclass").agg({"Age":'mean'})
#Mostramos el dataframe
edad_media
```

***12. Muestra por pantalla el porcentaje de menores y mayores de edad que sobrevivieron en cada clase.***

```python
#Agregamos la fila "Mayor de edad" con la clausula where de np
df_titanic['Mayor de edad'] = np.where(df_titanic['Age']>=18, 'Si', 'No')
#Creamos un dataframe filtrado 
mym = df_titanic.groupby(["Pclass","Mayor de edad"]).agg({"PassengerId":'count'})
mym["%"] = (mym['PassengerId'] / 
                  mym['PassengerId'].sum()) * 100
mym
```

```python
import matplotlib.pyplot as plt

mym["%"].plot(kind="bar")










```
