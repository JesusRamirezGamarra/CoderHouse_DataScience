```python
# Responder en los espacios ___
#1 ) Genera un range de 1 a N, recuerda que el ultimo es excluyente por lo que deberas aumentarle un +1
#2 ) Generar un condición donde el primer ciclo solo se sume un +1
#3 ) Sumar las últimos dos elementos de la lista para que así se cumpla la secuencia fibonacci. 
# Si queremos acceder al último elemento de una lista[-1], si queremos acceder al penúltimo elemento de una lista[-2] y así sucesivamente...

def fibonacci(a):
  respuesta = 0
  lista = [0]
  for i in range(0,a+1):
    if i == 0:
      respuesta = lista[0] + 1
    else:
      respuesta = lista[-1] + lista[-2]
    
    lista.append(respuesta)
    print(lista)   
  #return respuesta
  
```

## Reto Número 2
Integra en la función fibonacci un **while** en vez del **for** y que termine hasta que la respuesta final de la secuencia fibonacci sea mayor de 150.

```python
# Responder en los espacios ___
# 1) Genera un fibonacci donde la variable "a" sea menor que el resultado final del fibo y termine hasta ahí.

def fibonacci(a):
  respuesta = 0
  lista = [0]
  while respuesta < a:
    if respuesta == 0:
      respuesta = lista[-1] + 1
    else:
      respuesta = lista[-1] + lista[-2]
    
    if respuesta < a:
      lista.append(respuesta)
      print(lista)      
      #return respuesta
    else:
      break
     #return respuesta
```

```python
fibonacci(150)
```
