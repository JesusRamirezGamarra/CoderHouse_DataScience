```python


def RangoInfo(bins,csv):# Aquí definimos una función “RangoInfo”. 

    
    #calculando Rango(s) en funcion del bins a elegir
    #bins = 50
    dfmax = max(df_Empleado.Salary) 
    dfmin = min(df_Empleado.Salary)
    width = ( dfmax - dfmin ) / bins
    #width = (dfmax - dfmin ) / bins
    print (f"valor de cada rango : {width}")

    frecuencia_ = 0
    rangoNro = 1
    valorInferior_ = 0
    valorSuperior_ = 0
    NroElementoRango = 0
    df_Rango_ = []
    for i in range(1,bins + 1):
        valorInferior = dfmin + ((i-1) * width)
        valorSuperior = dfmin +  (i * width)
        df_Rango = df_Empleado[ (df_Empleado['Salary']>= valorInferior) & 
                                (df_Empleado['Salary']< valorSuperior  )] [ ['Salary']]
        frecuencia = df_Rango.loc[:,['Salary']].shape[0]
        #print (f"frecuencia :{frecuencia_} frecuencia Nueva :{frecuencia} ")
        #print (f"rango :{i}  ")
        
        if int(frecuencia) > frecuencia_ :
            frecuencia_ = frecuencia
            rangoNro = i
            valorInferior_ = valorInferior
            valorSuperior_ = valorSuperior
            NroElementoRango = df_Rango.loc[:,['Salary']].shape[0]
            df_Rango_ = df_Rango
            
    print( f"Rango {rangoNro} : [{valorInferior_} , {valorSuperior_}[" )
    print( f"existen : {NroElementoRango} empleados en este rango")
    
    res = Histograma(csv,bins,df_Rango_,valorInferior_,valorSuperior_) 
    print (res)
    
    return ''
```

```python
res = RangoInfo(50,'source/HRDataset_v14.csv') # probamos funcion que al ingresar el bins te dice cual es en rango con mayor cantidad de empleados
print (res)
```
