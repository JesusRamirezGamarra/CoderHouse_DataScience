### 2 )Corregir los target a solo dos valores distintos ‘<=50k’ y ‘>50k’

```python
DF.groupby('Target').count()
```

```python
pd.options.mode.chained_assignment = None # evitar la alerta producida x el replace re asignado a target.
DF['Target'] = DF['Target'].str.replace('<=50K.','<=50K', regex=True)
```

```python
DF['Target'] = DF['Target'].str.replace('>50K.','>50K', regex=True)
pd.options.mode.chained_assignment = 'warn'
```

```python
DF.groupby('Target').count()
```

### 3 )Calcular la proporción que hay entre personas con salarios >50k respecto del total para Entre personas de distinto sexo, Raza y Educacion.
### Por ejemplo cuál es el porcentaje de mujeres con un salario mayor a 50K y compararlo con el de los hombres. Lo mismo para personas de distintas razas y Educación

```python
DF_More50k = DF[DF.Target.str.contains ('>50K')]
DF_More50k
```

```python
TablaPivot=pd.pivot_table(
    DF_More50k,
    index=['Sex'],
    columns=['Target'],
    values=['ID'],
    aggfunc = lambda x : ( x.count()*100 / len(DF_More50k) ) 
     ) 
TablaPivot
```

```python
TablaPivot=pd.pivot_table(
    DF,
    index=['Sex'],
    columns=['Target'],
    values=['ID'],
    aggfunc = lambda x : ( x.count()*100 / len(DF) ) 
     ) 
TablaPivot
```

```python
# mostramos % en funcion al total de registros o considerando unicamente los registros con Target > 50k
```

```python
TablaPivot=pd.pivot_table(
    DF_More50k,
    index=['Race'],
    columns=['Target'],
    values=['ID'],
    aggfunc = lambda x : ( x.count()*100 / len(DF_More50k))
     ) 
TablaPivot
```
